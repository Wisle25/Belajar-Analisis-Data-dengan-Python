## **A. Pengantar Penerapan Dasar-Dasar Descriptive Statistics**

Statistik adalah ilmu yang menerapkan berbagai metode saintifik dalam pengumpulan, analisis, interpretasi, dan penyajian data. Dalam analisis data, statistik memegang peran penting, terutama **descriptive statistics**, yang digunakan untuk mendeskripsikan data melalui beberapa konsep utama:

### **Konsep Dasar Descriptive Statistics:**
1. **Measuring Central Tendency**: Mengidentifikasi nilai sentral dalam data.
2. **Measuring Dispersion**: Menilai variabilitas atau sebaran nilai dalam data.
3. **Measuring Asymmetry**: Menilai ketidaksimetrisan dalam distribusi data.
4. **Data Relationship**: Mengidentifikasi hubungan antar variabel dalam data.

### **Tipe Data**
Pemahaman terhadap tipe data sangat penting karena akan menentukan metode analisis dan bentuk visualisasi yang digunakan. Secara umum, data terbagi menjadi dua tipe:

#### **1. Data Kuantitatif (Numerik)**
Data yang direpresentasikan dalam skala numerik dan memungkinkan operasi matematis.

- **Data Kontinu**: Direpresentasikan dalam berbagai nilai numerik yang bisa dibagi atau diubah ke unit lebih kecil (contoh: umur, berat badan).
- **Data Diskret**: Direpresentasikan dalam bilangan bulat dan tidak bisa dibagi ke unit yang lebih kecil (contoh: jumlah kucing, jumlah produk terjual).

#### **2. Data Kategoris**
Data yang direpresentasikan dalam kelas atau label, biasanya berupa kategori.

- **Data Ordinal**: Data kategoris yang bisa diurutkan atau diberi peringkat (contoh: rating, kategori pelanggan).
- **Data Nominal**: Data kategoris yang tidak memiliki urutan atau peringkat (contoh: warna bulu, jenis kelamin).

---

### **Resume Singkat**
Descriptive statistics membantu mendeskripsikan data melalui konsep seperti central tendency, dispersion, asymmetry, dan data relationships. Memahami tipe data (kuantitatif atau kategoris) adalah kunci dalam memilih metode analisis yang tepat.


## **B. Measuring Central Tendency**

Measuring central tendency adalah sekumpulan parameter statistik yang digunakan untuk menggambarkan nilai khas atau sentral dari suatu data. Terdapat tiga parameter utama yang digunakan untuk ini, yaitu **mean**, **median**, dan **mode**.

Mari kita gunakan contoh data jumlah kucing peliharaan yang dimiliki pegawai Dicoding:

| **Nama Pegawai** | **Jumlah Kucing** |
|------------------|-------------------|
| Fajri            | 3                 |
| Hanifa           | 2                 |
| Iren             | 1                 |
| Hakim            | 1                 |
| Anggi            | 2                 |
| Muthia           | 3                 |
| Agis             | 2                 |
| Ilmi             | 1                 |
| Buchori          | 0                 |
| Zanuar           | 2                 |

### **1. Mean**
**Mean** atau rata-rata adalah nilai yang diperoleh dengan menjumlahkan seluruh data, kemudian membaginya dengan banyaknya data.

Contoh: Untuk mengetahui rata-rata jumlah kucing yang dimiliki pegawai Dicoding, kita hitung jumlah kucing total dan bagi dengan jumlah pegawai:

- **Jumlah total kucing**: 3 + 2 + 1 + 1 + 2 + 3 + 2 + 1 + 0 + 2 = 17
- **Jumlah pegawai**: 10
- **Mean**: 17 รท 10 = 1.7

Jadi, rata-rata jumlah kucing per pegawai adalah **1.7**.

Namun, perlu diperhatikan bahwa **mean** bisa sensitif terhadap outlier (nilai ekstrim). Jika misalnya ada kesalahan input di mana Hanifa memiliki 20 kucing, maka mean akan menjadi **3.5** dan tidak merepresentasikan kondisi nyata.

### **2. Median**
**Median** adalah nilai tengah dari data yang telah diurutkan. Median lebih tahan terhadap outlier karena hanya melihat data di tengah setelah diurutkan.

Langkah-langkah untuk menghitung median:
- Urutkan jumlah kucing: **0, 1, 1, 1, 2, 2, 2, 2, 3, 3**
- Karena jumlah data adalah genap (10), median adalah rata-rata dari dua nilai tengah: **(2 + 2) รท 2 = 2**

Jadi, median jumlah kucing adalah **2**.

### **3. Mode**
**Mode** adalah nilai yang paling sering muncul dalam kumpulan data.

Pada contoh data ini, nilai yang paling sering muncul adalah **2**, karena muncul sebanyak **4 kali**.

---

### **Resume Singkat**
Measuring central tendency memiliki tiga parameter utama:
- **Mean**: Rata-rata dari data. Dalam contoh, **1.7**, sensitif terhadap outlier.
- **Median**: Nilai tengah dari data yang diurutkan. Dalam contoh, **2**, tidak terpengaruh oleh outlier.
- **Mode**: Nilai yang paling sering muncul. Dalam contoh, **2**.


```python
import numpy as np
from scipy import stats

jumlah_kucing = np.array([3, 2, 1, 1, 2, 3, 2, 1, 0, 2])

# Menghitung mean dengan python np
print(jumlah_kucing.mean())

# Menghitung median dengan python np
print(np.median(jumlah_kucing))

# Menghitung mode dengan scipy dan np
mode_jumlah_kucing = stats.mode(jumlah_kucing)[0]
print(mode_jumlah_kucing)
```

## **C. Measuring Dispersion**

Measuring dispersion adalah konsep statistik yang digunakan untuk menggambarkan sebaran atau variasi dari suatu data. Dalam analisis data, selain mengetahui nilai sentral, kita juga perlu memahami bagaimana data tersebut tersebar. Ada empat parameter utama untuk mengukur dispersion, yaitu **range**, **interquartile range (IQR)**, **variance**, dan **standard deviation**.

Kita akan menggunakan data jumlah kucing peliharaan pegawai Dicoding sebagai contoh:

| **Nama Pegawai** | **Jumlah Kucing** |
|------------------|-------------------|
| Fajri            | 3                 |
| Hanifa           | 2                 |
| Iren             | 1                 |
| Hakim            | 1                 |
| Anggi            | 2                 |
| Muthia           | 3                 |
| Agis             | 2                 |
| Ilmi             | 1                 |
| Buchori          | 0                 |
| Zanuar           | 2                 |

### **1. Range**
**Range** adalah selisih antara nilai maksimum dan nilai minimum dalam data. Range memberikan gambaran kasar mengenai variasi data.

- **Nilai maksimum**: 3
- **Nilai minimum**: 0
- **Range**: 3 - 0 = 3

Range ini menunjukkan bahwa selisih antara jumlah kucing tertinggi dan terendah adalah 3.

### **2. Interquartile Range (IQR)**
**IQR** adalah selisih antara kuartil ketiga (Q3) dan kuartil pertama (Q1), yang menggambarkan rentang data di mana sebagian besar titik data berada.

- **Q1 (kuartil pertama)**: 1
- **Q3 (kuartil ketiga)**: 2
- **IQR**: 2 - 1 = 1

IQR menunjukkan bahwa sebagian besar data berkisar antara 1 dan 2.

### **3. Variance**
**Variance** mengukur seberapa jauh titik data menyimpang dari rata-rata (mean) dalam bentuk kuadrat selisih. Variance menunjukkan seberapa besar variasi yang ada dalam data.

- **Mean** dari jumlah kucing: 1.7
- Hitung selisih kuadrat dari mean untuk setiap titik data, lalu rata-rata selisih kuadrat tersebut menghasilkan variance.

Variance untuk data kucing pegawai Dicoding adalah **0.9**, yang berarti ada variasi kecil dalam jumlah kucing yang dimiliki.

### **4. Standard Deviation**
**Standard deviation** adalah akar kuadrat dari variance dan memberikan gambaran tentang simpangan data dalam unit yang sama dengan data aslinya. Standard deviation lebih mudah diinterpretasikan daripada variance karena memiliki satuan yang sama dengan data.

- **Standard deviation** dari data kucing pegawai Dicoding adalah **0.95**.

Semakin kecil nilai standard deviation, semakin dekat data ke nilai mean, dan semakin besar sebaran nilai data, semakin besar pula nilai standard deviation.

---

### **Summary (Variance vs. Standard Deviation)**:
- **Mean** adalah rata-rata dari semua elemen.
- **Variance** adalah seperti "FAKE" rata-rata perbedaan dari mean karena dihitung dalam satuan **kuadrat**, membuatnya sulit diinterpretasikan dalam konteks asli data.
- **Standard Deviation** adalah seperti "REAL" rata-rata perbedaan dari mean, karena satuannya dikembalikan ke satuan asli data (bukan kuadrat), sehingga lebih mudah dipahami.


```python
import pandas as pd

# Menghitung Range
range = jumlah_kucing.max() - jumlah_kucing.min()
print(range)

# Menghitung IQR
# NOTE: 75 adalah persentase untuk quartil ketiga dan 25 adalah persentase untuk quartil pertama
iqr = np.percentile(jumlah_kucing, 75) - np.percentile(jumlah_kucing, 25)
print(iqr)

# Menghitung Variance dan std deviation
jumlah_kucing_series = pd.Series(jumlah_kucing)
print(jumlah_kucing_series.var())
print(jumlah_kucing_series.std())
```

## **E. Data Relationship**

Data relationship adalah konsep statistik yang digunakan untuk mengidentifikasi hubungan antar **feature bertipe numerik** dalam suatu data. Ada dua parameter utama yang bisa digunakan untuk mengidentifikasi hubungan antar feature, yaitu **correlation** dan **covariance**.

### **1. Correlation**
![](https://dicoding-web-img.sgp1.cdn.digitaloceanspaces.com/original/academy/dos:54aa188e366ee32fb428a6938df3151820230309132752.png)
**Correlation** mengukur hubungan antara dua feature numerik. Nilai correlation berkisar antara **-1 hingga 1**:
- **Nilai positif** menunjukkan korelasi positif atau hubungan searah (positive correlation).
- **Nilai negatif** menunjukkan korelasi negatif atau hubungan berlawanan (negative correlation).
- **Nilai nol** berarti tidak ada korelasi (no correlation).

Misalnya, pada sebuah dataset, jika usia seseorang naik dan nilai keterampilan komunikasinya juga meningkat, itu berarti ada **positive correlation**. Jika satu meningkat sementara yang lain menurun, itu berarti ada **negative correlation**.

### **2. Covariance**
**Covariance** adalah parameter lain yang digunakan untuk mengukur hubungan antar dua feature. Sama seperti correlation, covariance juga dapat:
- **Positive covariance** berarti kedua feature berkorelasi positif (jika satu naik, yang lain naik).
- **Negative covariance** berarti kedua feature berkorelasi negatif (jika satu naik, yang lain turun).
- **Zero covariance** berarti tidak ada hubungan antara kedua feature.

Namun, nilai covariance tidak ternormalisasi, sehingga interpretasinya bisa lebih sulit dibandingkan correlation.

---

### **Pertanyaan: Apa Bedanya Correlation dan Covariance?**
**Covariance** dan **correlation** sama-sama mengukur hubungan antara dua variabel, tetapi ada perbedaan penting antara keduanya:

#### 1. **Covariance**:
- Mengukur **arah** hubungan (positif/negatif) tetapi **tidak distandarisasi**, jadi nilainya bergantung pada satuan data (misalnya dolar, kilometer).
- Covariance memberi kita informasi tentang apakah dua variabel bergerak **bersesuaian** (positif) atau **berlawanan** (negatif), tetapi sulit diinterpretasikan karena nilainya tidak dinormalkan.

#### 2. **Correlation**:
- Mengukur **arah dan kekuatan** hubungan dalam **skala standar** antara -1 dan 1.
- Correlation lebih mudah dipahami karena distandarisasi, sehingga kita bisa membandingkan korelasi antara dataset yang berbeda.

---

### **Mengapa Covariance Masih Digunakan jika Correlation Lebih Mudah?**
Meskipun correlation lebih mudah dipahami dan digunakan, covariance masih penting dalam beberapa kasus:

1. **Pengukuran Mentah**: Covariance memberikan pengukuran mentah tentang bagaimana dua variabel bergerak bersama dalam satuan asli (misalnya, dolar atau meter), yang berguna dalam analisis keuangan atau ketika satuan asli penting.
   
2. **Dasar dari Correlation**: Untuk menghitung correlation, kita perlu menghitung covariance terlebih dahulu, lalu menormalkannya dengan membagi dengan standar deviasi dari kedua variabel.
   
3. **Penggunaan Khusus**: Dalam bidang seperti keuangan, covariance digunakan untuk menilai hubungan mentah antara variabel seperti return investasi dalam satuan dolar.

### **Ringkasan**:
- **Covariance** memberikan ukuran **mentah, tidak dinormalkan** tentang bagaimana dua variabel bergerak bersama. Ini digunakan ketika kita perlu bekerja dengan data dalam **satuan aslinya**.
- **Correlation** memberikan versi hubungan yang **dinormalkan**, yang lebih mudah diinterpretasikan dan dibandingkan antar dataset yang berbeda.



```python
# Menggunakan library pandas untuk melihat korelasinya
sample_data = {
    'name': ['John', 'Alia', 'Ananya', 'Steve', 'Ben'],
    'age': [24, 22, 23, 25, 28],
    'communication_skill_score': [85, 70, 75, 90, 90],
    'quantitative_skill_score': [80, 90, 80, 75, 70]
}

df = pd.DataFrame(sample_data)
# Korelasi
df.corr(numeric_only=True)
```

```python
# Covariance
df.cov(numeric_only=True)
```
