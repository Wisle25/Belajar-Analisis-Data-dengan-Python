## **A. Pengenalan Analisis Data**

Analisis data adalah proses mengumpulkan, mentransformasi, dan menata data untuk menarik kesimpulan, membuat prediksi, atau mengambil keputusan yang lebih baik. Dalam kehidupan sehari-hari, kita sering menggunakan analisis data tanpa sadar, misalnya saat membeli barang secara online dengan melihat rating dan testimoni.

Seorang Data Analyst bekerja layaknya seorang detektif, yang mencari solusi menggunakan data melalui proses analisis. Proses ini membutuhkan beberapa keterampilan penting:

1. **Analytical Skills (Keterampilan Analitik):**

- Rasa ingin tahu yang tinggi.
- Kemampuan memahami konteks di balik data.
- Pemikiran teknis dan terorganisir.
- Mampu mengolah data dengan desain yang baik.
- Menerapkan strategi data dalam manajemen.

2. **Analytical Thinking (Pemikiran Analitik):**

- Mampu menyajikan data dalam bentuk visual untuk memudahkan pemahaman.
-Berpikir strategis untuk mencapai tujuan.
-Berorientasi pada masalah dan solusi.
-Mampu mengidentifikasi hubungan antara data.
-Menggunakan pemikiran besar (big-picture thinking) -dan perhatian pada detail.

---

### **Resume Singkat**

## **B. Pentingnya Data dalam Dunia Bisnis**

Data memainkan peran penting dalam dunia bisnis modern. Setiap perusahaan, mulai dari e-commerce hingga healthcare, menggunakan data untuk mengidentifikasi peluang, mengikuti tren, mengoptimalkan proses, dan meningkatkan layanan pelanggan. Hal ini menjadikan data sebagai "The new oil" karena sangat berharga untuk mendukung pengambilan keputusan yang tepat.

Salah satu pendekatan yang berkembang pesat di dunia bisnis adalah **data-driven decision making**, yaitu pengambilan keputusan berbasis data. Ini berarti perusahaan menggunakan data yang terorganisir dan diolah dengan baik untuk mendukung setiap langkah bisnis yang sesuai dengan tujuan mereka.

### **Karakteristik Perusahaan yang Menerapkan Data-Driven Decision Making:**
1. **Data tersentralisasi dan terorganisir**.
2. **Kebijakan tata kelola data** yang jelas.
3. **Tingkat aksesibilitas data yang tinggi** di seluruh organisasi.
4. **Penggunaan tool analisis** yang mudah dipakai.

### **Manfaat Data-Driven Decision Making:**
1. **Akuntabilitas dan konsistensi** dalam pengambilan keputusan.
2. **Mendorong inovasi** dan peluang bisnis baru.
3. **Meningkatkan efisiensi** dengan data yang relevan untuk membantu keputusan.

### **Cara Menerapkan Data-Driven Decision Making:**
1. **Mengumpulkan data sebanyak mungkin** dengan memastikan kualitasnya baik.
2. **Menentukan tujuan yang dapat diukur** untuk kemudahan evaluasi dan perbaikan.
3. **Memastikan aksesibilitas data** di seluruh bagian organisasi.
4. **Menentukan prioritas** dalam pengambilan keputusan untuk memaksimalkan dampak dan penggunaan sumber daya.

---

### **Resume Singkat**
Data sangat penting dalam dunia bisnis untuk mengidentifikasi tren, meningkatkan layanan, dan membuat keputusan berdasarkan bukti. **Data-driven decision making** memungkinkan perusahaan untuk membuat keputusan yang konsisten dan efisien dengan data yang tersentralisasi, meningkatkan inovasi, dan peluang. Untuk menerapkannya, diperlukan pengumpulan data yang baik, tujuan yang terukur, aksesibilitas data, dan penetapan prioritas.


## **C. Mengenal Siklus Analisis Data**

Analisis data adalah proses yang melibatkan beberapa tahapan untuk mengumpulkan, mentransformasi, dan menata data, hingga menghasilkan kesimpulan, prediksi, dan rekomendasi keputusan. Berikut adalah siklus umum yang digunakan dalam analisis data:

### **1. Mendefinisikan Pertanyaan**
Pada tahap ini, kita harus mendefinisikan pertanyaan analisis terkait data atau masalah yang ingin diselesaikan. Pertanyaan yang jelas membantu dalam memahami masalah bisnis dan memfokuskan pada data yang relevan. Konsep **4W** (What, Why, When, Where) sering digunakan dalam perumusan pertanyaan:
- **What (Apa):** Apa yang ingin diketahui dari data?
- **Why (Mengapa):** Mengapa pola atau tren tersebut muncul?
- **When (Kapan):** Kapan pola tersebut terjadi?
- **Where (Dimana):** Di mana peristiwa tersebut terjadi?

### **2. Data Wrangling**
Tahap ini mencakup pengumpulan dan pembersihan data. Data bisa berasal dari database internal, file eksternal, atau API. Setelah data terkumpul, kualitas dan struktur data harus dinilai, dan proses pembersihan dilakukan untuk memperbaiki atau menghapus data yang bermasalah. Proses ini sering memakan waktu paling banyak.

### **3. Exploratory Data Analysis (EDA)**
EDA adalah eksplorasi data untuk mendapatkan insight awal menggunakan **descriptive statistics**. Tujuannya adalah menemukan pola, hubungan, dan intuisi dari data. Selain descriptive statistics, machine learning atau inferential statistics juga dapat digunakan untuk prediksi di masa depan (meskipun di luar cakupan kelas ini).

### **4. Data Visualization**
Setelah EDA, temuan dari data divisualisasikan menggunakan berbagai teknik **visualisasi data**. Tujuannya adalah mengomunikasikan insight secara efektif.

### **5. Draw Conclusion & Communicate**
Tahap akhir adalah menarik kesimpulan dari analisis yang menjawab pertanyaan awal. Selain itu, hasil analisis harus dikomunikasikan, baik melalui laporan, presentasi, atau slide deck.

### **Catatan Penting:**
Siklus analisis data seringkali bersifat iteratif. Setelah EDA, bisa saja kita menemukan masalah baru yang membutuhkan pembersihan data tambahan, atau munculnya pola baru yang mengubah pertanyaan analisis awal. Inilah mengapa proses analisis data lebih cocok disebut sebagai siklus, bukan langkah linear.

---

### **Resume Singkat**
Siklus analisis data dimulai dari mendefinisikan pertanyaan, mengumpulkan dan membersihkan data (data wrangling), melakukan **Exploratory Data Analysis (EDA)** untuk menemukan pola, memvisualisasikan hasil, dan menarik kesimpulan. Siklus ini bersifat iteratif, sering membutuhkan perbaikan dan penyesuaian selama proses analisis.



## **D. Pengenalan Tool Pengolahan Data**

Dalam pengolahan data menggunakan Python, terdapat beberapa tools penting yang sering digunakan oleh para praktisi data. Python menjadi pilihan utama karena memiliki banyak **libraries** yang open-source dan mendukung proses analisis data.

Berikut adalah tiga libraries populer untuk pengolahan data dalam Python:

### **1. Library NumPy**
**NumPy** (numerical Python) adalah library powerful yang digunakan untuk membuat dan mengolah multi-dimensional arrays (matriks atau tensor). NumPy sangat berguna untuk komputasi numerik dengan berbagai fungsi seperti **basic linear algebra**, simulasi acak, dan **Fourier transforms**.

### **2. Library Pandas**
**Pandas** adalah library yang dirancang khusus untuk **manipulasi dan analisis data**. Pandas menggunakan data structure bernama **DataFrame**, yang mirip dengan tabel dalam basis data, untuk mengelola data secara efisien.

### **3. Library SciPy**
**SciPy** (Scientific Python) adalah library yang fokus pada **komputasi saintifik**, memanfaatkan NumPy sebagai komponen dasar. SciPy sering digunakan dalam bidang sains dan teknik untuk menyelesaikan perhitungan matematis yang kompleks, termasuk statistik.

Python memiliki tiga library utama untuk pengolahan data:
- **NumPy** untuk komputasi numerik dan pengolahan array.

- **SciPy** untuk komputasi saintifik dan statistik yang kompleks.

## **E. Pengenalan Tool Visualisasi Data**

Selain mengolah data, dalam proses analisis data kita juga perlu merepresentasikan data dalam bentuk visual yang menarik dan mudah dipahami. Python memiliki beberapa library open-source yang populer untuk membantu membuat visualisasi data. Berikut dua library yang paling sering digunakan oleh praktisi data:

### **1. Library Matplotlib**
**Matplotlib** adalah library Python yang komprehensif untuk membuat berbagai jenis visualisasi data, baik statis, animasi, maupun interaktif. Library ini menyediakan banyak objek dan metode untuk membantu kita memodifikasi tampilan visual sesuai kebutuhan.

Matplotlib memiliki submodule **matplotlib.pyplot**, sering disingkat sebagai **plt**, yang menyediakan fungsi untuk membuat visualisasi data di atas kanvas (figure). Selain itu, plt juga menawarkan opsi untuk mendefinisikan jenis visualisasi dan menambahkan elemen dekoratif seperti label, judul, dll.

### **2. Library Seaborn**
**Seaborn** adalah library Python yang dirancang khusus untuk membuat visualisasi data yang menarik dan informatif. Seaborn menggunakan matplotlib di belakang layar untuk menghasilkan grafik yang lebih kompleks dan estetis.

Seaborn sangat berguna dalam merepresentasikan data dengan cara yang lebih **atraktif dan informatif**, sehingga sering dipadukan dengan matplotlib untuk menghasilkan visualisasi yang lebih kaya dan interaktif.

---

### **Resume Singkat**
Untuk visualisasi data di Python, dua library yang paling sering digunakan adalah:
- **Matplotlib**: menyediakan alat yang komprehensif untuk membuat visualisasi data yang dapat disesuaikan dengan kebutuhan.
- **Seaborn**: membantu membuat visualisasi yang lebih estetis dan informatif, sering dipadukan dengan matplotlib untuk hasil yang lebih kaya.

Kedua library ini merupakan fondasi utama yang akan digunakan dalam membuat visualisasi data yang menarik di dunia industri.
